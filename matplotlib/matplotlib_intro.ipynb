## Two Interfaces for the Price of One
#### MATLAB-style interface

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

#### Object-oriented interface

```python
# First create a grid of plots
# ax sill be an aray of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Simple Line Plots

```python
plt.style.use('seaborn-whitegrid')
```

```python
# start by creating a figure and axes
fig = plt.figure()
ax = plt.axes()
```

 - **Figure** can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels.
 - **Axes** is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.

```python
# use ax.plot to plot some data
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
# We can also use the pylab interface and let figure and axes be created in background
plt.plot(x, np.sin(x));
```

```python
# Create a single figure with multiple lines by calling plot multiple times
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse') # all HTML color names supported
```

```python
# adjusting line styles
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
# can combine linestyle and color in one go
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

## Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# if you want either axis to be displayed in reverse
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# can also use plt.axis[xmin, xmax, ymin, ymax]
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
# or tighten the bounds
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# Or ensuring equal aspect ratio
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)');
```

```python
# bring it all together
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

### Matplotlib Gotchas

MATLAB-style functions vs. object-oriented methods
 - `plt.xlabel() -> ax.set_xlabel()`
 - `plt.ylabel() -> ax.set_ylabel()`
 - `plt.xlim() -> ax.set_xlim()`
 - `plt.ylim() -> ax.set_ylim()`
 - `plt.title() -> ax.set_title()`

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2), xlabel='x', ylabel='sin(x)', title='A Simple Plot');
```
